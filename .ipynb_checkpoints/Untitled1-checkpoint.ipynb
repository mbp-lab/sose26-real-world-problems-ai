```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Fix random seed for reproducibility
np.random.seed(42)

# Parameters
n_features = 100
true_feature_idx = 0   # index of the feature that actually influences the outcome
noise_std = 0.2

# Different dataset sizes to explore
sample_sizes = [20, 50, 100, 500, 1000]

for n_samples in sample_sizes:
    print(f"\n=== Dataset size: {n_samples} ===")

    # 1. Generate features (independent normal distribution)
    X = np.random.randn(n_samples, n_features)

    # 2. Generate ground truth (only one feature matters)
    true_feature = X[:, true_feature_idx]
    y = true_feature + np.random.randn(n_samples) * noise_std

    # 3. Train linear regression
    model = LinearRegression()
    model.fit(X, y)

    # Extract coefficients
    coefs = model.coef_

    # Get top 5 features by absolute magnitude
    top_indices = np.argsort(np.abs(coefs))[-5:][::-1]

    print("Top 5 features by coefficient magnitude:")
    for idx in top_indices:
        print(f"Feature {idx}: coefficient = {coefs[idx]:.4f}")

    # 4. Visualization
    plt.figure()
    plt.bar(range(5), coefs[top_indices])
    plt.xticks(range(5), [f"F{idx}" for idx in top_indices])
    plt.title(f"Top 5 coefficients (n={n_samples})")
    plt.ylabel("Coefficient value")
    plt.xlabel("Feature")
    plt.show()
```
