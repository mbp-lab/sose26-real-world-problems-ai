##### Collecting new data is often time- and cost-intensive. So let's entertain an alternative: instead of gathering more samples, can we get more information on the sample we already have? Think of a medical scenario where for each patient we can get additional information, such as records from their previous check-ups or measurements collected during routine tests. Let's simulate such a scenario.

Synthetic patient data - Generate a dataset with a fixed number of patients (N=50) and vary the number of recorded variables F (e.g. 10-200, in increments). Draw the variables independently from a normal distribution.

Generate ground truth (clinical outcome) - Assume that only a small subset of these variables (5 out of F) actually influences the clinical outcome (e.g. disease severity score). Select 5 variables and use them to generate the target variable (hint:linear combinations). Add some noise (e.g. Gaussian noise with standard deviation of 0.1) to simulate measurement noise and unobserved factors.

Write down your intuitions - You will next train a regression model to predict the clinical outcome from the patient data you created. What do you expect: how will model performance change as more patient variables are added?

Train a regression model using different number of variables F. Run the model 50 times and visualize the model performance for different values of F.

Describe what you see.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

N = 50
F_list = np.arange(10, 201, 5)
t_f = np.array([0, 1, 2, 3, 4])
w = np.array([1.0, -0.5, 0.3, 0.2, 0.1])
noise_std = 0.1
n_runs = 50

mean_r2_train = []
mean_r2_val = []
mean_mse_train = []
mean_mse_val = []
mean_mae_train = []
mean_mae_val = []
k = 0
for F in F_list:
    r2_tr = []
    r2_val = []
    mse_tr = []
    mse_va = []
    mae_tr = []
    mae_va = []
    for run in range(n_runs):
        np.random.seed(k)
        k += 1
        X = np.random.randn(N, F)
        y = X[:, t_f] @ w + noise_std * np.random.randn(N)
        X_tr, X_va, y_tr, y_va = train_test_split(X, y, test_size=0.2, random_state=run)
        model = LinearRegression()
        model.fit(X_tr, y_tr)
        pred_tr = model.predict(X_tr)
        pred_va = model.predict(X_va)
        r2_tr.append(r2_score(y_tr, pred_tr))
        r2_val.append(r2_score(y_va, pred_va))
        mse_tr.append(mean_squared_error(y_tr, pred_tr))
        mse_va.append(mean_squared_error(y_va, pred_va))
        mae_tr.append(mean_absolute_error(y_tr, pred_tr))
        mae_va.append(mean_absolute_error(y_va, pred_va))
    mean_r2_train.append(sum(r2_tr) / n_runs)
    mean_r2_val.append(sum(r2_val) / n_runs)
    mean_mse_train.append(sum(mse_tr) / n_runs)
    mean_mse_val.append(sum(mse_va) / n_runs)
    mean_mae_train.append(sum(mae_tr) / n_runs)
    mean_mae_val.append(sum(mae_va) / n_runs)

plt.figure()
plt.plot(F_list, mean_r2_train, marker="o", ms=3, label="train $R^2$ (mean)")
plt.plot(F_list, mean_r2_val, marker="o", ms=3, label="validation $R^2$ (mean)")
plt.xlabel("number of variables F")
plt.ylabel(r"$R^2$")
plt.legend()
plt.title("N = 50 patients, noise_std = {}, 50 runs per F".format(noise_std))
plt.show()

plt.figure()
plt.plot(F_list, mean_mse_train, marker="o", ms=3, label="train MSE (mean)")
plt.plot(F_list, mean_mse_val, marker="o", ms=3, label="validation MSE (mean)")
plt.xlabel("number of variables F")
plt.ylabel("MSE")
plt.legend()
plt.title("Mean squared error (mean over 50 runs)")
plt.show()

plt.figure()
plt.plot(F_list, mean_mae_train, marker="o", ms=3, label="train MAE (mean)")
plt.plot(F_list, mean_mae_val, marker="o", ms=3, label="validation MAE (mean)")
plt.xlabel("number of variables F")
plt.ylabel("MAE")
plt.legend()
plt.title("Loss: mean absolute error (mean over 50 runs)")
plt.show()
```

- Take the most interesting number of variable that you identified in Station 2.
- Now, keep the number of variables fixed to what you identified as "the most interesting" and vary the sample size so that the maximum N = F_interesting
- For each N, fit linear regression and **on the same axes** compare **ground-truth coefficients** (before / generating weights on the 5 signal features) with **OLS coefficients after regression** on those same features. Use one **shared horizontal scale** across all N so panels are directly comparable.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Set F_interesting to the number of variables you chose in Station 2.
F_interesting = 200

t_f = np.array([0, 1])
w = np.array([1.0, -0.5])
noise_std = 0.1

N_min = 50
N_values = np.unique(np.append(np.arange(N_min, F_interesting, 50), F_interesting))

rows = []
for N in N_values:
    np.random.seed(42_000 + N)
    X = np.random.randn(N, F_interesting)
    y = X[:, t_f] @ w + noise_std * np.random.randn(N)
    model = LinearRegression()
    model.fit(X, y)
    coef_hat = model.coef_
    true_b = w
    est_b = coef_hat[t_f]
    rows.append((N, true_b, est_b))

all_vals = np.concatenate([np.r_[r[1], r[2]] for r in rows])
vmax = np.max(np.abs(all_vals))
x_lim = (-vmax * 1.05, vmax * 1.05)

fig, axes = plt.subplots(len(rows), 1, figsize=(8, 1.8 * len(rows)), sharex=True, constrained_layout=True)
if len(rows) == 1:
    axes = np.array([axes])

y_feat = np.arange(len(t_f))
h = 0.35
labels = ["$x_{{{}}}$".format(int(j) + 1) for j in t_f]

for ax, (N, true_b, est_b) in zip(axes, rows):
    ax.barh(y_feat - h / 2, true_b, height=h, color="C1", label="before (ground truth)")
    ax.barh(y_feat + h / 2, est_b, height=h, color="C0", label="after regression")
    ax.set_yticks(y_feat)
    ax.set_yticklabels(labels)
    ax.set_ylabel("$N = {}$".format(N))
    ax.set_xlim(x_lim)
    ax.axvline(0, color="k", linewidth=0.6)

axes[0].legend(loc="upper right", fontsize=8)
axes[-1].set_xlabel(r"coefficient value (shared scale across all $N$)")
fig.suptitle(
    r"Signal-feature coefficients: ground truth vs. OLS (fixed $F = %s$)" % F_interesting,
    fontsize=11,
)
plt.show()
```
