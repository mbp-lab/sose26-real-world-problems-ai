- Synthetic data - Generate datasets with a fixed sample size (N=50) but varying number of features F (e.g. 10-200 in increments). Draw the features independently from a normal distribution.

- Generate ground truth - From the feature set F, select 5 features to generate the target variable (hint:linear combinations). This way only a subset of features will carry signal. Add some noise (e.g. Gaussian noise with standard deviation of 0.1) to your ground truth to simulate real world noisy data.

- Write down your intuitions - You will next train a regression model to predict the ground truths of the synthetic data you created. What do you think: how would each dataset perform?

- Train a regression model - use each dataset you created to predict the associated ground truth. Run the model 10 times and visualize the performance.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split

N = 50
F_list = np.arange(10, 201, 5)
t_f = np.array([0, 1, 2, 3, 4])
w = np.array([1.0, -0.5, 0.3, 0.2, 0.1])
noise_std = 0.1
n_runs = 10

mean_r2_train = []
mean_r2_val = []
k = 0
for F in F_list:
    r2_tr = []
    r2_val = []
    for run in range(n_runs):
        np.random.seed(k)
        k += 1
        X = np.random.randn(N, F)
        y = X[:, t_f] @ w + noise_std * np.random.randn(N)
        X_tr, X_va, y_tr, y_va = train_test_split(X, y, test_size=0.2, random_state=run)
        model = LinearRegression()
        model.fit(X_tr, y_tr)
        r2_tr.append(r2_score(y_tr, model.predict(X_tr)))
        r2_val.append(r2_score(y_va, model.predict(X_va)))
    mean_r2_train.append(sum(r2_tr) / n_runs)
    mean_r2_val.append(sum(r2_val) / n_runs)

plt.figure()
plt.plot(F_list, mean_r2_train, marker="o", ms=3, label="train $R^2$ (mean)")
plt.plot(F_list, mean_r2_val, marker="o", ms=3, label="validation $R^2$ (mean)")
plt.xlabel("number of features F")
plt.ylabel(r"$R^2$")
plt.legend()
plt.title("N = 50, noise_std = {}, 10 runs per F".format(noise_std))
plt.show()
```
