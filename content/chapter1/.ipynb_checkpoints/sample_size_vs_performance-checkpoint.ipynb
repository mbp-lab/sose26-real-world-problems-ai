- Generate synthetic data - create datasets with a fixed number of features F = 50 but varying sample size N (e.g. from 10 to 200 in increments). Draw all features independently from a normal distribution.

- Generate ground truth - From the feature set F, select 5 features to generate the target variable. This way only a subset of features will carry signal.

- Train a regression model - use each dataset you created to predict the associated ground truth. Visualize the performance of your model.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

F = 50
N_list = np.arange(10, 201, 5)
t_f = np.array([0, 1, 2, 3, 4])
w = np.array([1.0, -0.5, 0.3, 0.2, 0.1])

r2 = []
for i, N in enumerate(N_list):
    np.random.seed(i)
    X = np.random.randn(N, F)
    y = X[:, t_f] @ w
    model = LinearRegression()
    model.fit(X, y)
    r2.append(r2_score(y, model.predict(X)))

plt.figure()
plt.plot(N_list, r2, marker="o", ms=3)
plt.xlabel("sample size N")
plt.ylabel(r"$R^2$")
plt.title("F = 50")
plt.show()
```
