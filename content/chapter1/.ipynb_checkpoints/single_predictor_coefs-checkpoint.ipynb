# One predictive feature: coefficient plots across sample sizes

- Sample sizes **N ∈ {100, 250, 500, 750, 1000}**.
- Number of features **F ∈ {10, 50}**. Only **one** feature (index 0) appears in the **ground-truth** target:  
  **y = β₀ x₀ + ε** with β₀ = 1, **ε ~ N(0, σ²)**, all columns of **X** still drawn i.i.d. **N(0,1)**.
- **OLS** and **Ridge** are fit on the full **X**. For each **N**, we show the **top k** coordinates by **|**OLS coefficient|** and, on the same axes, **ground truth**, **OLS**, and **Ridge** (shared **x**-scale across panels in each figure).

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Ridge

N_values = [100, 250, 500, 750, 1000]
F_settings = [10, 50]
signal_idx = 0
beta_signal = 1.0
noise_std = 0.1
top_k = 5
ridge_alpha = 1.0
rng_base = 42_000


def run_panels(F, N_values, rng_base):
    true_beta = np.zeros(F)
    true_beta[signal_idx] = beta_signal
    rows = []
    for j, N in enumerate(N_values):
        np.random.seed(rng_base + j * 1000 + F)
        X = np.random.randn(N, F)
        y = X[:, signal_idx] * beta_signal + noise_std * np.random.randn(N)
        ols = LinearRegression()
        ols.fit(X, y)
        coef_ols = ols.coef_
        k_show = min(top_k, F)
        idx = np.argsort(-np.abs(coef_ols))[:k_show]
        ridge = Ridge(alpha=ridge_alpha)
        ridge.fit(X, y)
        coef_ridge = ridge.coef_
        v_true = true_beta[idx]
        v_ols = coef_ols[idx]
        v_ridge = coef_ridge[idx]
        rows.append((N, idx, v_true, v_ols, v_ridge))
    all_vals = np.concatenate([np.r_[r[2], r[3], r[4]] for r in rows])
    vmax = np.max(np.abs(all_vals))
    x_lim = (-vmax * 1.05, vmax * 1.05)
    fig, axes = plt.subplots(
        len(rows), 1, figsize=(8, 2.0 * len(rows)), sharex=True, constrained_layout=True
    )
    if len(rows) == 1:
        axes = np.array([axes])
    bar_h = 0.22
    dy = 0.28
    for ax, (N, idx, v_true, v_ols, v_ridge) in zip(axes, rows):
        order = np.argsort(-np.abs(v_ols))
        idx_s = idx[order]
        gt = v_true[order]
        a = v_ols[order]
        b = v_ridge[order]
        y_pos = np.arange(len(idx_s))
        ax.barh(y_pos - dy, gt, height=bar_h, color="C1", label="ground truth")
        ax.barh(y_pos, a, height=bar_h, color="C0", label="OLS")
        ax.barh(y_pos + dy, b, height=bar_h, color="C2", label="Ridge ($\\alpha=%s$)" % ridge_alpha)
        ax.set_yticks(y_pos)
        ax.set_yticklabels(["$x_{{{}}}$".format(int(i) + 1) for i in idx_s])
        ax.set_ylabel("$N = {}$".format(N))
        ax.set_xlim(x_lim)
        ax.axvline(0, color="k", linewidth=0.6)
    axes[0].legend(loc="upper right", fontsize=8)
    axes[-1].set_xlabel(r"coefficient (shared scale across $N$ in this panel)")
    fig.suptitle(
        r"Top %d by $|\hat{\beta}_{\mathrm{OLS}}|$ — truth, OLS, Ridge ($F=%d$, single signal: $x_{%d}$)"
        % (top_k, F, signal_idx + 1),
        fontsize=11,
    )
    plt.show()


for F in F_settings:
    run_panels(F, N_values, rng_base)
```
