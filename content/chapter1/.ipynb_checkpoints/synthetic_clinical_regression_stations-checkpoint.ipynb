# Synthetic patient data: varying features (Station 2) and sample size (follow-on)

This notebook follows the written steps only: fixed $N=100$, varying number of variables $F$; then fixed $F=100$, varying $N$.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error

rng = np.random.default_rng(42)
```

## Part A — Synthetic patient data (fixed $N=100$, vary $F$)

**Step 1.** Generate a dataset with a fixed number of patients **$N=100$** and vary the number of recorded variables **$F$** (e.g. 10–200, in increments). Draw the variables **independently from a normal distribution**.

```python
N = 100
F_values = np.arange(10, 201, 10)  # 10, 20, ..., 200


def draw_patient_matrix(n_patients: int, n_features: int, random_state: np.random.Generator) -> np.ndarray:
    """Each entry i.i.d. N(0, 1)."""
    return random_state.standard_normal(size=(n_patients, n_features))
```

**Step 2.** Generate ground truth (clinical outcome): only **5** of the $F$ variables influence the outcome. **Select 5 variables** and generate the target as a **linear combination**. Add **Gaussian noise** with standard deviation **0.1**.

```python
NOISE_STD = 0.1
# Causal features are the first five columns (indices 0..4); present for all F >= 5.
CAUSAL_IDX = np.array([0, 1, 2, 3, 4])
true_weights = np.array([1.5, -0.8, 0.6, -1.1, 0.4])  # linear combination weights on those 5 columns


def generate_outcome(X: np.ndarray, random_state: np.random.Generator) -> np.ndarray:
    """y = sum_k w_k * X[:, k] + eps, eps ~ N(0, NOISE_STD^2), k in CAUSAL_IDX."""
    linear_part = X[:, CAUSAL_IDX] @ true_weights
    noise = random_state.normal(loc=0.0, scale=NOISE_STD, size=X.shape[0])
    return linear_part + noise
```

**Step 3.** *Write down your intuitions.* You will train a regression model to predict the clinical outcome from the patient data. **What do you expect: how will model performance change as more patient variables are added? Why?**

**Intuitions (to fill in your own words):**

- As $F$ grows while $N$ stays fixed, the model gets **more inputs**, including many **non-causal** (pure noise) variables. Ordinary least squares can **fit spurious correlations** in the training set, especially when $F$ is large relative to $N$.
- I therefore expect **performance on held-out data** to **worsen** as irrelevant features accumulate (higher variance / overfitting), unless the procedure strongly limits flexibility. Performance on training data might look **misleadingly good** when $F$ is large.
- The **signal** (only 5 true predictors) stays the same; extra columns mostly add **noise** for estimation.

**Step 4.** Train a **regression model** using different numbers of variables $F$. **Run the model 100 times** and **visualize** model performance for different values of $F$.

Each run: regenerate $X$ and $y$, hold out 20% for testing, fit `LinearRegression` on the train split, record **test $R^2$** and **test RMSE**.

```python
N_REPEATS = 100
test_r2_runs = {F: [] for F in F_values}
test_rmse_runs = {F: [] for F in F_values}

for F in F_values:
    for rep in range(N_REPEATS):
        rs = np.random.default_rng(rep * 10_000 + F)
        X = draw_patient_matrix(N, F, rs)
        y = generate_outcome(X, rs)
        X_train, X_test, y_train, y_test = train_test_split(
            X, y, test_size=0.2, random_state=rep
        )
        model = LinearRegression()
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        test_r2_runs[F].append(r2_score(y_test, y_pred))
        test_rmse_runs[F].append(np.sqrt(mean_squared_error(y_test, y_pred)))

mean_r2 = np.array([np.mean(test_r2_runs[F]) for F in F_values])
std_r2 = np.array([np.std(test_r2_runs[F], ddof=1) for F in F_values])
mean_rmse = np.array([np.mean(test_rmse_runs[F]) for F in F_values])
std_rmse = np.array([np.std(test_rmse_runs[F], ddof=1) for F in F_values])

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].errorbar(F_values, mean_r2, yerr=std_r2, fmt="-o", capsize=3)
axes[0].set_xlabel("Number of variables F")
axes[0].set_ylabel("Test R² (mean ± std over 100 runs)")
axes[0].set_title("Linear regression: N=100 fixed")
axes[0].grid(True, alpha=0.3)

axes[1].errorbar(F_values, mean_rmse, yerr=std_rmse, fmt="-o", color="C1", capsize=3)
axes[1].set_xlabel("Number of variables F")
axes[1].set_ylabel("Test RMSE (mean ± std over 100 runs)")
axes[1].set_title("Same experiment: RMSE")
axes[1].grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

**Step 5.** *Describe what you see. Did your assumptions get confirmed? Do you notice anything surprising?*

**Observations:**

- Test $R^2$ tends to **decrease** as $F$ increases (with $N$ fixed), and test RMSE tends to **increase**: extra non-informative features give the model more room to **overfit** the training split.
- When $F$ approaches or exceeds $N$, the linear model is **highly flexible** relative to sample size; held-out performance often **deteriorates** even though the true outcome only depends on 5 variables.
- **Surprising** only if one expected “more data columns always helps”: here columns are **not** more labeled information—they are mostly **irrelevant predictors**, which **hurt** generalization under OLS with fixed $N$.
- This **matches** the intuition that irrelevant variables and a fixed sample size together worsen **out-of-sample** regression performance.

## Part B — Fixed $F=100$, varying $N$; one causal feature

- Fix **$F=100$**.
- Create datasets with **$N \in \{50, 75, 100, 125, 150\}$**.
- **One** feature defines the ground truth (plus Gaussian noise with $\sigma=0.1$ as in Part A).
- For each $N$, train **linear regression**, extract **learned coefficients**, and **visualize the top 5 features by magnitude** ($|w_j|$).
- Reflect: **How does sample size influence the coefficients?**

```python
F_fixed = 100
N_list = [50, 75, 100, 125, 150]
CAUSAL_FEATURE_INDEX = 0  # single causal column
TRUE_SLOPE = 2.0


def generate_outcome_single_feature(X: np.ndarray, random_state: np.random.Generator) -> np.ndarray:
    linear_part = TRUE_SLOPE * X[:, CAUSAL_FEATURE_INDEX]
    noise = random_state.normal(loc=0.0, scale=NOISE_STD, size=X.shape[0])
    return linear_part + noise


coef_by_N = {}
for N_train in N_list:
    rs = np.random.default_rng(7_000 + N_train)
    X = draw_patient_matrix(N_train, F_fixed, rs)
    y = generate_outcome_single_feature(X, rs)
    lr = LinearRegression()
    lr.fit(X, y)
    coef_by_N[N_train] = lr.coef_.copy()

# Top 5 features by |coefficient| for each N (feature indices may differ per N)
fig, axes = plt.subplots(1, len(N_list), figsize=(14, 4), sharey=True)
for ax, N_train in zip(axes, N_list):
    w = coef_by_N[N_train]
    top5_idx = np.argsort(np.abs(w))[-5:][::-1]
    ax.bar(np.arange(5), w[top5_idx], tick_label=[str(j) for j in top5_idx])
    ax.set_title(f"N = {N_train}")
    ax.set_xlabel("Feature index (top 5 by |coef|)")
    ax.tick_params(axis="x", rotation=45)
axes[0].set_ylabel("Learned coefficient")
plt.suptitle("Top 5 coefficients by magnitude (F=100, one true feature: index 0)", y=1.02)
plt.tight_layout()
plt.show()

# Overlay: coefficient on the true causal feature vs N
causal_estimates = [coef_by_N[n][CAUSAL_FEATURE_INDEX] for n in N_list]
plt.figure(figsize=(6, 4))
plt.axhline(TRUE_SLOPE, color="gray", linestyle="--", label=f"True slope ({TRUE_SLOPE})")
plt.plot(N_list, causal_estimates, "-o")
plt.xlabel("Sample size N")
plt.ylabel(f"Learned coef for feature {CAUSAL_FEATURE_INDEX}")
plt.title("Estimate of the causal feature vs sample size")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

**What do you observe? How does sample size influence the coefficients?**

- With **small $N$**, the estimated **top-5** coefficients often include **wrong** features (large $|w_j|$ on non-causal columns): finite-sample noise lets irrelevant predictors **correlate** with $y$.
- As **$N$ increases**, the estimate for the **true** feature (index 0) typically **stabilizes** closer to **2.0**, and the **magnitude ranking** more often puts the causal feature **among the largest** coefficients.
- **Larger sample size** reduces **variance** of coefficient estimates in this linear setting; **irrelevant** features’ coefficients tend to **shrink toward zero** relative to spurious spikes seen for small $N$.
