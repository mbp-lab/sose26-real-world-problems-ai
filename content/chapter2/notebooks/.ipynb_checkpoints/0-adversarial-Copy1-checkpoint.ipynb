Station 1 provided a nice conceptual warm-up. However, you aren't required to manually create adversarial attacks for this exercise but can use an attack method available in Python. In the next cell, you can already see some code to guide you. Note that the code has been generated via an LLM but checked for the purposes of this exercise. In that step, additional comments have been added to aid your unterstanding! 

Note that you need to install libraries, such as sklearn, xgboost or art to be able to use the code!


```python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from xgboost import XGBClassifier
from art.estimators.classification import XGBoostClassifier as ART_XGBoostClassifier
from art.attacks.evasion import BoundaryAttack

# Load dataset
# 0 = malignant, 1 = benign
data = load_breast_cancer()
X = data.data.astype(np.float32)
y = data.target.astype(np.int64)
print("Dataset shape:", X.shape)

# Train-test split
# stratify=y preserves class balance
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, stratify=y, random_state=42)


# Scale features
# StandardScaler makes features comparable:
# mean = 0, std = 1
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train).astype(np.float32)
X_test = scaler.transform(X_test).astype(np.float32)

# ART requires valid feature ranges
clip_values = (np.min(X_train, axis=0),np.max(X_train, axis=0))

# Train XGBoost model - some standard parameters
xgb = XGBClassifier(
    n_estimators=100,     # number of trees
    max_depth=3,          # tree depth
    learning_rate=0.05,  # smaller = more stable learning
    eval_metric="logloss",
    random_state=42)
xgb.fit(X_train, y_train)
clean_acc = accuracy_score(y_test, xgb.predict(X_test))
print("Accuracy on original data:", round(clean_acc, 4))


# Wrap model with ART
# ART attacks cannot directly use normal XGBoost models.
# The wrapper acts as an adapter between: XGBoost model <-> ART attack library
art_xgb = ART_XGBoostClassifier(
    model=xgb,
    clip_values=clip_values,
    nb_features=X_train.shape[1],
    nb_classes=2)
X_attack = X_test#[:20] # BoundaryAttack is slow, so use only a few samples if you run into runtime issues
y_attack = y_test#[:20]


# Create untargeted BoundaryAttack. Untargeted means that any wrong prediction is counted as a legitimate attack, we don't target a specific class
attack = BoundaryAttack(
    estimator=art_xgb,
    targeted=False,  
    max_iter=20,
    num_trial=20,
    sample_size=20,
    verbose=False)

# Generate adversarial examples
print("\nRunning BoundaryAttack...")
X_adv = attack.generate(x=X_attack)
# Get predictions - ART predict() returns probabilities/logits- np.argmax converts them into class labels.
clean_preds = np.argmax(art_xgb.predict(X_attack), axis=1)
adv_preds = np.argmax(art_xgb.predict(X_adv), axis=1)
print("Done!")
```

After evaluating standard adversarial attacks, the next step is to explore adversarial training as a defense mechanism. The idea behind adversarial training is that the model is exposed to adversarial examples during training so that it learns to become more robust against attacks.

```python
# Final robustness evaluation

art_xgb_retrained = ART_XGBoostClassifier(
    model=xgb_retrained,
    clip_values=clip_values,
    nb_features=X_train.shape[1],
    nb_classes=2
)


def art_predict_labels(model, X):
    return np.argmax(model.predict(X), axis=1)


def attack_success_rate(model, X_original, X_adversarial):
    pred_original = art_predict_labels(model, X_original)
    pred_adversarial = art_predict_labels(model, X_adversarial)
    return np.mean(pred_original != pred_adversarial)


def evaluate_model(model, X_clean, y_true, X_adv):

    clean_preds = art_predict_labels(model, X_clean)
    adv_preds = art_predict_labels(model, X_adv)

    return {
        "Clean accuracy": accuracy_score(y_true, clean_preds),
        "Adversarial accuracy": accuracy_score(y_true, adv_preds),
        "Attack success rate": np.mean(clean_preds != adv_preds)
    }


# Attack against original model
attack_original = BoundaryAttack(
    estimator=art_xgb,
    targeted=False,
    max_iter=20,
    num_trial=20,
    sample_size=20,
    verbose=False
)

# Attack against retrained model
attack_retrained = BoundaryAttack(
    estimator=art_xgb_retrained,
    targeted=False,
    max_iter=20,
    num_trial=20,
    sample_size=20,
    verbose=False
)


print("\nGenerating adversarial test samples against original model...")
X_adv_from_original = attack_original.generate(x=X_attack)

print("Generating adversarial test samples against retrained model...")
X_adv_from_retrained = attack_retrained.generate(x=X_attack)


# 2x2 robustness matrix
results_2x2 = pd.DataFrame(
    {
        "Evaluated on original model": [
            attack_success_rate(
                art_xgb,
                X_attack,
                X_adv_from_original
            ),
            attack_success_rate(
                art_xgb,
                X_attack,
                X_adv_from_retrained
            )
        ],

        "Evaluated on retrained model": [
            attack_success_rate(
                art_xgb_retrained,
                X_attack,
                X_adv_from_original
            ),
            attack_success_rate(
                art_xgb_retrained,
                X_attack,
                X_adv_from_retrained
            )
        ]
    },

    index=[
        "Attack generated against original model",
        "Attack generated against retrained model"
    ]
)

print("\nAttack success rate matrix")
display(results_2x2)


# Full evaluation table
results_eval = pd.DataFrame(
    {
        "Original model on adversarial samples from original model": evaluate_model(
            art_xgb,
            X_attack,
            y_attack,
            X_adv_from_original
        ),

        "Original model on adversarial samples from retrained model": evaluate_model(
            art_xgb,
            X_attack,
            y_attack,
            X_adv_from_retrained
        ),

        "Retrained model on adversarial samples from original model": evaluate_model(
            art_xgb_retrained,
            X_attack,
            y_attack,
            X_adv_from_original
        ),

        "Retrained model on adversarial samples from retrained model": evaluate_model(
            art_xgb_retrained,
            X_attack,
            y_attack,
            X_adv_from_retrained
        )
    }
).T

print("\nDetailed robustness evaluation")
display(results_eval)
```
