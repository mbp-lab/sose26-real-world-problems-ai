#### 🟢 Station 1 - Autism classification accuracy

Machine learning is widely used in biomedical research and healthcare. One prominent example is autism classification, where researchers use a variety of data sources, such as brain imaging or eye-tracking data, to distinguish between people with and without autism. These studies are well known for their small sample sizes, an issue we encountered in Chapter 1. In this station, we will examine the reported accuracies of 55 studies that used different methods for autism detection.

````python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import spearmanr

# Load the CSV file
df = pd.read_csv("../data/autism-classification-accuracy.csv")

# Calculate Spearman correlation
corr, p_value = spearmanr(df["Sample size"], df["Accuracy"])

# Scatter plot
plt.figure(figsize=(8, 5))
plt.scatter(df["Sample size"], df["Accuracy"])

# Labels and title
plt.xlabel("Sample Size")
plt.ylabel("Accuracy")
plt.title(f"Sample Size vs Accuracy\nSpearman r = {corr:.3f}, p = {p_value:.3g}")

# Show grid
plt.grid(True)

# Display the plot
plt.show()

# Print correlation values
print(f"Spearman correlation: {corr:.3f}")
print(f"P-value: {p_value:.3g}")
````

The result challenges the simple expectation that a larger study should automatically report a higher classification accuracy. Instead, the scatter plot shows a clear negative relationship between sample size and reported accuracy (Spearman r = -0.669, p = 2.44e-08): smaller autism classification studies tend to report much higher accuracies, while larger studies tend to report more modest results.

A likely explanation is that small-sample studies are more vulnerable to overfitting, optimistic validation, and sampling noise. With few participants, it is easier for a model to pick up accidental patterns that do not generalize to new data. Publication bias may also contribute: very small studies with disappointing results are less likely to be published, while unusually high accuracies are more noticeable. Methodological choices such as feature selection before validation, weak train/test separation, repeated model tuning on the same data, or lack of external validation could also inflate accuracy.

#### 🟢🟢 Station 2 - In Folds We Trust...


To make efficient use of limited data, validation techniques are often used to evaluate models while retaining as much data as possible for training. One widely used technique is k-fold cross-validation. In the words of scikit learn: " In k-fold CV, the training set is split into k smaller sets. The following procedure is followed for each of the k “folds”. A model is trained using k-1 of the folds as training data; the resulting model is validated on the remaining part of the data (i.e., it is used as a test set to compute a performance measure such as accuracy). The performance measure reported by k-fold cross-validation is then the average of the values computed in the loop." 

Below is a code snippet that applies SVM to a binary classification problem. The model performance is evaluated using both a holdout split and 10-fold cross-validation.

Explore the code and describe the results below. What stands out to you?

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.svm import SVC
from sklearn.feature_selection import SelectKBest,f_classif
from sklearn.model_selection import (train_test_split,StratifiedKFold,GridSearchCV)

# Load dataset
df = pd.read_csv("../data/Pareidolia_dataset.csv")

X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values

sample_size = 50
n_repeats = 50
n_features_selected = 20 # # Number of highest-scoring features (ANOVA F-test) to keep

# C controls regularization strength; larger C = fits training data more aggressively
# gamma controls boundary smoothness; larger gamma = more localized / complex boundary
param_grid = {"C": [0.01, 0.1, 1, 10, 100, 1000],
    "gamma": [0.0001, 0.001, 0.01, 0.1, 1, 10]}

holdout_scores = []
cv_scores = []
rng = np.random.default_rng(0)

for i in range(n_repeats):
    # Random subset
    idx = rng.choice(len(y), size=sample_size, replace=False)
    X_subset = X[idx]
    y_subset = y[idx]

    # -------------------------
    # Holdout evaluation
    # -------------------------

    X_train, X_test, y_train, y_test = train_test_split(
        X_subset,
        y_subset,
        test_size=0.2,
        stratify=y_subset,
        random_state=i)

    # Keep top-scoring features
    selector = SelectKBest(f_classif,k=n_features_selected)
    X_train_selected = selector.fit_transform(X_train,y_train)
    X_test_selected = selector.transform(X_test)

    holdout_model = SVC(kernel="rbf",C=1.0,gamma=0.01)
    holdout_model.fit(X_train_selected,y_train)
    holdout_scores.append(
        holdout_model.score(X_test_selected,y_test))

    # -------------------------
    # Cross-validation
    # -------------------------

    # Select top-scoring features
    selector = SelectKBest(f_classif,k=n_features_selected)
    X_selected = selector.fit_transform(X_subset,y_subset)

    # Hyperparameter search
    search = GridSearchCV(estimator=SVC(kernel="rbf"),param_grid=param_grid,cv=5,scoring="accuracy")
    search.fit(X_selected,y_subset)
    best_params = search.best_params_

    cv_model = SVC(kernel="rbf",C=best_params["C"],gamma=best_params["gamma"])
    cv = StratifiedKFold(n_splits=10,shuffle=True,random_state=i)

    fold_scores = []
    for train_idx,test_idx in cv.split(X_selected,y_subset):

        X_train_cv = X_selected[train_idx]
        X_test_cv = X_selected[test_idx]

        y_train_cv = y_subset[train_idx]
        y_test_cv = y_subset[test_idx]

        cv_model.fit(X_train_cv,y_train_cv)

        fold_scores.append(cv_model.score(X_test_cv,y_test_cv))

    cv_scores.append(np.mean(fold_scores))

# Results
mean_holdout = np.mean(holdout_scores)
mean_cv = np.mean(cv_scores)

print("Average accuracy")
print("----------------")
print(f"Holdout:    {mean_holdout:.3f}")
print(f"10-fold CV: {mean_cv:.3f}")

# Plot
plt.figure(figsize=(5,4))
plt.bar(["Holdout","10-fold CV"],[mean_holdout,mean_cv])
plt.axhline(0.5,linestyle="--",color="black",label="Chance")
plt.ylabel("Average accuracy")
plt.ylim(0,1)
plt.legend()
plt.tight_layout()
plt.show()
````

### 💡 Summarize your main insights here: what stands out to you so far?  - 

What stands out is the large gap between the two reported estimates: the simple holdout evaluation is close to chance accuracy (about 0.510), whereas the 10-fold cross-validation estimate is much higher (about 0.812). At first glance, this makes cross-validation look much better, but the difference is suspiciously large.

The important issue is that the cross-validation pipeline is not properly nested. Feature selection is performed once on the full subset before the cross-validation folds are created, so information from the future test folds leaks into the training process. Hyperparameter tuning is also done on the same selected data before the final 10-fold evaluation, which can make the CV estimate overly optimistic. In contrast, the holdout approach keeps the test data separate during feature selection, so it gives a more realistic estimate here.

The main lesson is that cross-validation is not automatically trustworthy. Every data-dependent step, including feature selection, scaling, dimensionality reduction, and hyperparameter tuning, must be fitted only on the training portion inside each fold. Otherwise, the model indirectly sees the test data and the reported accuracy can become inflated.

A correct analysis pipeline should look like this:

```text
Full dataset
   │
   ├── Repeat evaluation / outer split
   │      │
   │      ├── Outer training data
   │      │      │
   │      │      └── Inner cross-validation for model selection
   │      │              For each inner fold:
   │      │              1. Fit preprocessing / feature selection on inner-training fold only
   │      │              2. Transform inner-validation fold with the fitted preprocessing
   │      │              3. Train SVM with one hyperparameter setting
   │      │              4. Evaluate on inner-validation fold
   │      │
   │      │      Choose best hyperparameters from inner-CV
   │      │      Refit preprocessing + feature selection on the full outer-training data
   │      │      Train final model on the transformed outer-training data
   │      │
   │      └── Outer test data
   │             Transform using preprocessing fitted on outer-training data only
   │             Evaluate final model once
   │
   └── Average outer-test scores across repeats/folds
```

Key rule: the test fold must stay completely untouched until the final evaluation step. Any operation that learns from the data must be inside the training side of the split.
