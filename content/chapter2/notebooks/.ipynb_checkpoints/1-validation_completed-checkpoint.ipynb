#### 🟢🟢 Station 1 - In Folds We Trust...

Machine learning models need data for learning but also for model evaluation. In limited data scenarios,  K-Fold cross-validation is economical as it enables using all the data for training and also to reuse all of it for validation. In the words of scikit-learn: "In k-fold CV, the training set is split into k smaller sets. The following procedure is followed for each of the k “folds”. A model is trained using k-1 of the folds as training data; the resulting model is validated on the remaining part of the data (i.e., it is used as a test set to compute a performance measure such as accuracy). The performance measure reported by k-fold cross-validation is then the average of the values computed in the loop." You can see the visualization of k-fold from scikit-learn on the left in the image below. 

![Caption](../images/k-fold.png)


The figure on the right shows a case where SVM was applied to a binary classification problem. The model performance is evaluated using both a holdout split and 10-fold cross-validation for a sample size of 100. 


### 💡 Summarize your main insights below: How does the use of 10-fold CV affect the model performance? Why? What additional information would you want to know about the model behind the figure on the right?

My main insight is that 10-fold cross-validation should usually give a more stable and less split-dependent estimate of model performance than one holdout split. In the example with a sample size of 100, the single holdout result can change strongly depending on which observations happen to land in the test set. With 10-fold CV, every observation is used for validation once and for training nine times, so the final score averages over ten validation results rather than relying on one lucky or unlucky split.

This does **not** automatically mean the model becomes better. Cross-validation mainly improves the **estimate of generalization performance** and makes better use of limited data. It can also reduce the variance of the performance estimate, but it can still be optimistically biased if preprocessing, feature selection, model selection, or hyperparameter tuning are performed before the folds are created.

- Were preprocessing and feature selection performed separately inside each training fold?

```python
# Station 2 - Autism classification accuracy
# This cell loads the dataset if it is available locally and explores the
# relationship between sample size and reported classification accuracy.

import os
import glob
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def find_autism_accuracy_file(search_roots=(".", "..", "../..", "/mnt/data")):
    """Find a likely Autism Classification Accuracy data file."""
    patterns = []
    for root in search_roots:
        for ext in ("*.csv", "*.xlsx", "*.xls", "*.tsv"):
            patterns.append(str(Path(root) / "**" / ext))
    candidates = []
    for pattern in patterns:
        candidates.extend(glob.glob(pattern, recursive=True))
    keywords = ("autism", "classification", "accuracy")
    scored = []
    for file in candidates:
        name = Path(file).name.lower()
        score = sum(k in name for k in keywords)
        if score:
            scored.append((score, file))
    scored = sorted(scored, reverse=True)
    return scored[0][1] if scored else None

def load_table(path):
    path = Path(path)
    if path.suffix.lower() == ".csv":
        return pd.read_csv(path)
    if path.suffix.lower() == ".tsv":
        return pd.read_csv(path, sep="\t")
    if path.suffix.lower() in [".xlsx", ".xls"]:
        return pd.read_excel(path)
    raise ValueError(f"Unsupported file type: {path.suffix}")

def pick_column(df, possible_names, contains=None):
    lower_cols = {c.lower().strip(): c for c in df.columns}
    for name in possible_names:
        if name.lower() in lower_cols:
            return lower_cols[name.lower()]
    if contains:
        for c in df.columns:
            lc = c.lower()
            if all(term in lc for term in contains):
                return c
    return None

data_path = find_autism_accuracy_file()
if data_path is None:
    print("Dataset not found in this environment.")
    print("Put the 'Autism Classification Accuracy' csv/xlsx file in the notebook folder or /mnt/data, then rerun this cell.")
else:
    print(f"Loaded: {data_path}")
    df = load_table(data_path)
    display(df.head())
    print("\nShape:", df.shape)
    print("\nColumns:", list(df.columns))

    sample_col = pick_column(
        df,
        ["sample_size", "sample size", "n", "N", "total sample", "total_n", "participants"],
        contains=["sample"]
    )
    acc_col = pick_column(
        df,
        ["accuracy", "classification accuracy", "acc", "reported accuracy"],
        contains=["accuracy"]
    )

    if sample_col is None or acc_col is None:
        print("\nCould not automatically identify the sample-size and accuracy columns.")
        print("Please set sample_col and acc_col manually from the printed column names.")
    else:
        plot_df = df[[sample_col, acc_col]].copy()
        plot_df.columns = ["sample_size", "accuracy"]
        plot_df["sample_size"] = pd.to_numeric(plot_df["sample_size"], errors="coerce")
        plot_df["accuracy"] = pd.to_numeric(plot_df["accuracy"], errors="coerce")
        plot_df = plot_df.dropna()
        display(plot_df.describe())

        pearson = plot_df["sample_size"].corr(plot_df["accuracy"], method="pearson")
        spearman = plot_df["sample_size"].corr(plot_df["accuracy"], method="spearman")
        print(f"Pearson correlation:  {pearson:.3f}")
        print(f"Spearman correlation: {spearman:.3f}")

        plt.figure(figsize=(7, 5))
        plt.scatter(plot_df["sample_size"], plot_df["accuracy"], alpha=0.75)
        plt.xscale("log")
        plt.xlabel("Sample size (log scale)")
        plt.ylabel("Reported classification accuracy")
        plt.title("Autism classification studies: sample size vs. accuracy")
        plt.grid(True, alpha=0.3)
        plt.show()

        median_n = plot_df["sample_size"].median()
        plot_df["sample_size_group"] = np.where(
            plot_df["sample_size"] <= median_n, "smaller studies", "larger studies"
        )
        display(plot_df.groupby("sample_size_group")["accuracy"].agg(["count", "mean", "median", "std"]))
```

### 💡 Summarize your main insights below:

Were your expectations confirmed or challenged? If your expectations were confirmed, explain the reasoning behind them. If the results differed from your expectations, discuss possible explanations for the discrepancy. Think about aspects that are related to ML methodology as well as broader research factors. 


Before checking the data, I expected larger studies to show **lower but more reliable** classification performance. Small studies are more vulnerable to sampling noise, overfitting, unstable train/test splits, and publication bias, so they can produce very high reported accuracies that may not generalize.

The expected pattern is therefore not necessarily “larger sample size = better accuracy.” In biomedical ML, larger and more heterogeneous samples often make the task harder because they include more realistic variation across participants, scanners, sites, symptom profiles, and measurement conditions. A large sample can therefore reduce inflated performance estimates while giving a more trustworthy estimate.

If the plot and correlations show that small studies report the highest accuracies, that would support the idea that very high accuracy in small biomedical datasets should be interpreted cautiously. Possible explanations include overfitting, leakage from preprocessing or feature selection before validation, non-independent observations, repeated use of the same benchmark datasets, selective reporting of successful models, and publication bias. If larger studies do not clearly outperform smaller ones, this does not mean sample size is unimportant; it means reported accuracy is shaped by study design, validation quality, data heterogeneity, and reporting practices as well as by model quality.

```python
# Station 4 - Data leakage checks
# This cell explores the dataset for non-independent rows and repeated data sources.
# These are common leakage risks if rows are randomly split as if they were independent.

from pathlib import Path
import pandas as pd

if "df" not in globals():
    data_path = find_autism_accuracy_file()
    if data_path is None:
        print("Dataset not found. Please add the Autism Classification Accuracy file and rerun Station 2 first.")
        df = None
    else:
        df = load_table(data_path)

if df is not None:
    print("Shape:", df.shape)
    display(df.head())
    print("\nMissing values per column:")
    display(df.isna().sum().sort_values(ascending=False))

    duplicated_rows = df[df.duplicated(keep=False)]
    print(f"\nExact duplicated rows: {len(duplicated_rows)}")
    if len(duplicated_rows):
        display(duplicated_rows)

    id_like_cols = [
        c for c in df.columns
        if any(term in c.lower() for term in ["doi", "title", "study", "paper", "author", "reference", "citation"])
    ]
    print("\nPossible publication/study identifier columns:", id_like_cols)

    for col in id_like_cols:
        repeated = df[col].astype(str).str.strip().value_counts()
        repeated = repeated[repeated > 1]
        if len(repeated):
            print(f"\nRepeated values in '{col}':")
            display(repeated.head(20))

    dataset_like_cols = [
        c for c in df.columns
        if any(term in c.lower() for term in ["dataset", "data set", "database", "cohort", "source", "sample"])
    ]
    print("\nPossible dataset/source columns:", dataset_like_cols)

    for col in dataset_like_cols:
        repeated = df[col].astype(str).str.strip().value_counts()
        repeated = repeated[repeated > 1]
        if len(repeated):
            print(f"\nRepeated values in '{col}':")
            display(repeated.head(20))

    result_terms = ["accuracy", "acc", "auc", "sensitivity", "specificity", "precision", "recall", "f1"]
    non_result_cols = [c for c in df.columns if not any(t in c.lower() for t in result_terms)]
    if len(non_result_cols) >= 2:
        near_duplicates = df[df.duplicated(subset=non_result_cols, keep=False)]
        print(f"\nRows duplicated after ignoring result/performance columns: {len(near_duplicates)}")
        if len(near_duplicates):
            display(near_duplicates.sort_values(non_result_cols[:1]))

    print("\nInterpretation:")
    print("Leakage risk 1: repeated rows or repeated publications mean observations are not independent.")
    print("Leakage risk 2: repeated datasets/cohorts mean a random row split can put information from the same source in both train and test.")
    print("A safer approach is grouped splitting, for example grouping by paper/study ID or dataset/cohort.")
```

### 💡 Summarize your main insights from Station 4 here:

The two main leakage risks I would flag are:

1. **Non-independent rows from the same study or publication.** If one paper contributes several accuracy values, model variants, feature sets, or validation results, those rows should not be treated as independent observations. A random split could place results from the same study in both training and test data.

2. **Repeated use of the same dataset/cohort across rows.** Many autism classification papers reuse common datasets such as benchmark imaging or screening datasets. If rows based on the same participant pool or data source are split across train and test sets, the test set is no longer fully independent.

Both issues create overly optimistic estimates because the model can indirectly learn study-specific or dataset-specific patterns. A better strategy is to use grouped validation, where all rows from the same paper and/or same dataset are kept in the same fold. If the goal is to generalize to new studies, the most realistic evaluation would be leave-one-study-out or leave-one-dataset-out validation.
