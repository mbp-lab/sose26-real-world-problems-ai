### Model certainty on COMPAS

We train two binary classifiers on the COMPAS recidivism task and compare how **accuracy** relates to **confidence** (the model's predicted probability for the chosen class).

A well-calibrated model should follow the diagonal: when it says it is 80% confident, it should be correct about 80% of the time.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler

df = pd.read_csv("../data/compas_clean.csv")
df = df[df["race"].isin(["African-American", "Caucasian"])].copy()

features = ["sex", "age", "priors_count", "c_charge_degree"]
X = df[features]
y = df["two_year_recid"].astype(int)

train_idx, test_idx = train_test_split(
    df.index, test_size=0.25, random_state=42, stratify=y
)
X_train, X_test = X.loc[train_idx], X.loc[test_idx]
y_train, y_test = y.loc[train_idx], y.loc[test_idx]

cat_cols = X.select_dtypes(include="object").columns.tolist()
num_cols = [c for c in features if c not in cat_cols]

preprocessor = ColumnTransformer(
    [
        ("cat", OneHotEncoder(handle_unknown="ignore"), cat_cols),
        ("num", StandardScaler(), num_cols),
    ]
)

logistic_model = Pipeline(
    [
        ("prep", preprocessor),
        ("clf", LogisticRegression(max_iter=1000)),
    ]
)

nn_model = Pipeline(
    [
        ("prep", preprocessor),
        (
            "clf",
            MLPClassifier(
                hidden_layer_sizes=(64, 32),
                activation="relu",
                max_iter=500,
                random_state=42,
            ),
        ),
    ]
)

logistic_model.fit(X_train, y_train)
nn_model.fit(X_train, y_train)

models = {
    "Logistic regression": logistic_model,
    "Neural network": nn_model,
}

for name, model in models.items():
    y_pred = model.predict(X_test)
    print(f"{name} test accuracy: {accuracy_score(y_test, y_pred):.3f}")
```

```python
def confidence_scores(y_proba):
    y_pred = (y_proba >= 0.5).astype(int)
    return np.where(y_pred == 1, y_proba, 1 - y_proba), y_pred


def accuracy_by_confidence(y_true, y_proba, n_bins=10):
    confidence, y_pred = confidence_scores(y_proba)
    bin_edges = np.linspace(0.5, 1.0, n_bins + 1)

    centers, accuracies, counts = [], [], []
    for low, high in zip(bin_edges[:-1], bin_edges[1:]):
        if high == bin_edges[-1]:
            mask = (confidence >= low) & (confidence <= high)
        else:
            mask = (confidence >= low) & (confidence < high)
        if mask.sum() == 0:
            continue
        centers.append((low + high) / 2)
        accuracies.append((y_pred[mask] == y_true[mask]).mean())
        counts.append(mask.sum())

    return np.array(centers), np.array(accuracies), np.array(counts)


fig, ax = plt.subplots(figsize=(8, 6))

for name, model in models.items():
    y_proba = model.predict_proba(X_test)[:, 1]
    centers, accuracies, counts = accuracy_by_confidence(y_test, y_proba)
    ax.plot(centers, accuracies, marker="o", linewidth=2, label=name)

ax.plot([0.5, 1], [0.5, 1], "--", color="gray", label="Perfect calibration")
ax.set_xlabel("Confidence score")
ax.set_ylabel("Accuracy")
ax.set_title("Accuracy vs confidence on COMPAS test set")
ax.set_xlim(0.5, 1.0)
ax.set_ylim(0.5, 1.0)
ax.legend()
ax.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
